```python
import pandas as pd
```

```python
netflix_df = pd.read_csv("./total_netflix_2023 new.csv")

netflix_df.head(10)
```

```python
netflix_df["Release Date"] = pd.to_datetime(netflix_df["Release Date"],errors='coerce')
netflix_df["Hours Viewed"] = pd.to_numeric(netflix_df["Hours Viewed"], errors='coerce')
netflix_df['Rating'] = pd.to_numeric(netflix_df['Rating'], errors='coerce')
netflix_df['Number of Ratings'] = pd.to_numeric(netflix_df['Number of Ratings'], errors='coerce')

# Filtra as linhas onde a conversão para número foi bem-sucedida
netflix_df = netflix_df[pd.notna(netflix_df["Release Date"])]
netflix_df = netflix_df[pd.notna(netflix_df['Number of Ratings'])]


# Exibe o DataFrame resultante
print(netflix_df.info())
```

```python
netflix_df[["Genre"]]
```

```python
def limpar_genre(valor):
    if valor == "0" or valor == "Genre":
        return "Unknown"
    if isinstance(valor, str):
        # Remove as aspas e colchetes
        valor = valor.replace("'", "").replace("[", "").replace("]", "")
        # Divide os gêneros e remove espaços em branco
        generos = [g.strip() for g in valor.split(',')]
        # Concatena os gêneros de volta em uma string
        return ','.join(generos)
    else:
        return valor


def serapar_genre(valor,num):

    lista = valor.split(",")
            
    if len(lista) == 1:
        lista.append(None)
        lista.append(None)
    elif len(lista) == 2:
        lista.append(None)

    return lista[num]
```

```python
# Aplica a função à coluna "Genre"
netflix_df["Genre"] = netflix_df["Genre"].apply(limpar_genre)

# Aplica a função Separar_genre
for i in range(3):
    texto = f'Genre_{i}'  
    netflix_df[texto] = netflix_df["Genre"].apply(serapar_genre, num = i)
```
